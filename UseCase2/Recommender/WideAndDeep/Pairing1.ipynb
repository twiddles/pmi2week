```python
#This is a very slow pairing search


#First select one part of all the data
shortAth06 = Ath06[:513200]

#Ath06[:513200]
#Ath06[513200:1026400]
#Ath06[1026400:1539600]
#Ath06[1539600:]



#l = [a.index(r[2]) for r in shortAth06.itertuples()]

#Now loop through all the authors and find the index location of their name in the array a 

#array l contains at the end indexes at which given author names have been found in a

#Some names do not match and this throws an exception in index search. In that case, we write "-1" and count an error.
#About 1% of data had errors (non-matches)

countErrors = 0
l = []
for r in shortAth06.itertuples():
    try:
        l.append(a.index(r[2]))
    except ValueError:
        l.append(-1)
        countErrors += 1
        print("Error: ", countErrors, r[0], r[2])



print("done")
```
